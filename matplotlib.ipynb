```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Plot 1: Line Plot
x = np.linspace(0, 5, 100)
y1 = np.sin(x)
plt.figure(figsize=(8, 5))
plt.plot(x, y1, label='sin(x)')
plt.title('Line Plot - sin(x)')
plt.legend()
plt.show()
```

```python
# Plot 2: Bar Plot
categories = ['A', 'B', 'C', 'D']
values = [7, 3, 9, 5]
plt.figure(figsize=(8, 5))
plt.bar(categories, values, color='orange')
plt.title('Bar Plot - Categorical Data')
plt.show()
```

```python
# Plot 3: Histogram
data = np.random.normal(0, 1, 1000)
plt.figure(figsize=(8, 5))
plt.hist(data, bins=30, color='green', alpha=0.7)
plt.title('Histogram - Random Data')
plt.show()
```

```python
# Plot 4: Pie Chart
sizes = [25, 35, 20, 20]
labels = ['Apple', 'Banana', 'Orange', 'Grapes']
plt.figure(figsize=(8, 5))
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title('Pie Chart - Fruit Distribution')
plt.show()
```

```python
# Plot 5: Scatter Plot
x = np.random.randn(100)
y = np.random.randn(100)
colors = np.random.rand(100)
sizes = 100 * np.random.rand(100)
plt.figure(figsize=(8, 5))
plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='viridis')
plt.colorbar()
plt.title('Scatter Plot - Random Data')
plt.show()
```

```python
# Plot 6: Stacked Bar Plot
N = 5
men_means = [20, 35, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]
ind = np.arange(N)
width = 0.35
plt.figure(figsize=(8, 5))
p1 = plt.bar(ind, men_means, width)
p2 = plt.bar(ind, women_means, width, bottom=men_means)
plt.ylabel('Scores')
plt.title('Stacked Bar Plot - Scores by group and gender')
plt.xticks(ind, ('G1', 'G2', 'G3', 'G4', 'G5'))
plt.legend((p1[0], p2[0]), ('Men', 'Women'))
plt.show()
```

```python
# Plot 7: Box Plot
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
plt.figure(figsize=(8, 5))
plt.boxplot(data, vert=True, patch_artist=True)
plt.title('Box Plot - Random Data')
plt.show()
```

```python
# Plot 8: Violin Plot
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
plt.figure(figsize=(8, 5))
plt.violinplot(data)
plt.title('Violin Plot - Random Data')
plt.show()
```

```python
# Plot 9: Area Plot
x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
plt.figure(figsize=(8, 5))
plt.fill_between(x, y1, y2, color='skyblue', alpha=0.4)
plt.title('Area Plot - Sin(x) and Cos(x)')
plt.show()
```

```python
labels = ['A', 'B', 'C', 'D']
men_means = [20, 35, 30, 35]
women_means = [25, 32, 34, 20]

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, men_means, width, label='Men')
rects2 = ax.bar(x + width/2, women_means, width, label='Women')

# Add some text for labels, title, and custom x-axis tick labels, etc.
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

plt.show()
```

```python
# Subplots
x = np.linspace(0, 5, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(10, 6))

plt.subplot(2, 1, 1)
plt.plot(x, y1)
plt.title('Sine Wave')

plt.subplot(2, 1, 2)
plt.plot(x, y2)
plt.title('Cosine Wave')

plt.tight_layout()
plt.show()
```

```python
# Customized Line Plots
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y1, label='sin(x)', color='blue', linestyle='--')
plt.plot(x, y2, label='cos(x)', color='red', linestyle='-.')
plt.title('Customized Line Plots')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid()
plt.show()
```

```python
# Error Bars
x = np.linspace(0, 10, 50)
y = np.sin(x)
error = np.random.rand(50) / 2
plt.figure(figsize=(8, 6))
plt.errorbar(x, y, yerr=error, fmt='o', color='green')
plt.title('Error Bars')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid()
plt.show()
```

```python
# Contour Plot
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(X) + np.cos(Y)

plt.figure(figsize=(8, 6))
plt.contourf(X, Y, Z, cmap='viridis')
plt.colorbar()
plt.title('Contour Plot')
plt.show()
```

```python
# Plot 10: 3D Plot
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(8, 5))
ax = fig.add_subplot(111, projection='3d')
x = np.random.standard_normal(100)
y = np.random.standard_normal(100)
z = np.random.standard_normal(100)
ax.scatter(x, y, z, c='r', marker='o')
ax.set_title('3D Scatter Plot')
plt.show()
```

```python
# 3D Surface Plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
X = np.linspace(-5, 5, 100)
Y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2))

ax.plot_surface(X, Y, Z, cmap='plasma')
ax.set_title('3D Surface Plot')
plt.show()
```
